```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and the specific sheet into a DataFrame
file_path = 'ken_admpop_2019.xlsx'
sheet_name = 'ken_admpop_ADM2_2019'
df_adm2 = pd.read_excel(file_path, sheet_name=sheet_name)

# Define the correct columns for Gen Z (ages 7-22 in 2019)
gen_z_columns = ['T_05_09', 'T_10_14', 'T_15_19', 'T_20_24']

# Sum the Gen Z population for each region
df_adm2['Gen_Z_Total'] = df_adm2[gen_z_columns].sum(axis=1)

# Display the DataFrame with the Gen Z total population
print(df_adm2[['ADM0_NAME', 'ADM1_NAME', 'ADM2_NAME', 'Gen_Z_Total']])

# Plot the population distribution of Gen Z
plt.figure(figsize=(10, 6))
df_adm2.groupby('ADM2_NAME')['Gen_Z_Total'].sum().sort_values().plot(kind='barh')
plt.xlabel('Gen Z Population')
plt.ylabel('Region')
plt.title('Population Distribution of Gen Z by Region')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file and the specific sheet into a DataFrame
file_path = 'ken_admpop_2019.xlsx'
sheet_name = 'ken_admpop_ADM1_2019'
df_adm1 = pd.read_excel(file_path, sheet_name=sheet_name)

# Define the correct columns for Gen Z (ages 7-22 in 2019)
gen_z_columns = ['T_05_09', 'T_10_14', 'T_15_19', 'T_20_24']

# Sum the Gen Z population for each region
df_adm1['Gen_Z_Total'] = df_adm1[gen_z_columns].sum(axis=1)

# Define the correct columns for Gen Z by gender
gen_z_female_columns = ['F_05_09', 'F_10_14', 'F_15_19', 'F_20_24']
gen_z_male_columns = ['M_05_09', 'M_10_14', 'M_15_19', 'M_20_24']

# Sum the Gen Z population by gender for each region
df_adm1['Gen_Z_Female'] = df_adm1[gen_z_female_columns].sum(axis=1)
df_adm1['Gen_Z_Male'] = df_adm1[gen_z_male_columns].sum(axis=1)

# Display the DataFrame with the Gen Z gender distribution
print(df_adm1[['ADM0_NAME', 'ADM1_NAME', 'Gen_Z_Female', 'Gen_Z_Male']])

# Get the top 5 counties with the highest number of Gen Z population
top_5_counties = df_adm1.nlargest(5, 'Gen_Z_Total')

# Plot the gender distribution of Gen Z
ax = df_adm1[['ADM1_NAME', 'Gen_Z_Female', 'Gen_Z_Male']].set_index('ADM1_NAME').plot(kind='bar', stacked=True, figsize=(14, 8), color=['#FFA07A', '#20B2AA'])
for idx, row in top_5_counties.iterrows():
    ax.get_children()[idx].set_color('#FF6347')  # Highlight Female bars
    ax.get_children()[idx + len(df_adm1)].set_color('#4682B4')  # Highlight Male bars

plt.xlabel('Region')
plt.ylabel('Population')
plt.title('Gender Distribution of Gen Z by Region (ADM1 Level) with Top 5 Counties Highlighted')
plt.legend(title='Gender')
plt.show()
```
