```python
import pandas as pd

# Load the Excel file
file_path = 'kdhs-tables.xlsx'
xls = pd.ExcelFile(file_path)

# List all sheet names
sheet_names = xls.sheet_names
print("Sheet Names:", sheet_names)

# Function to get column names and a brief description
def describe_sheet(sheet_name):
    df = pd.read_excel(xls, sheet_name=sheet_name)
    print(f"\nSheet: {sheet_name}")
    print("Columns and Description:")
    for col in df.columns:
        print(f"{col}: {df[col].dtype}")

# Iterate over all sheets to get columns and their descriptions
for sheet in sheet_names:
    describe_sheet(sheet)
```
