```python
# Convert observation_date to datetime
df['observation_date'] = pd.to_datetime(df['observation_date'])

# Define adjusted generation ranges to reflect age 18-25 years
generations = {
    "Silent Generation": ("1946-01-01", "1963-12-31"),
    "Baby Boomers": ("1964-01-01", "1982-12-31"),
    "Gen X": ("1983-01-01", "1998-12-31"),
    "Millennials": ("1999-01-01", "2014-12-31"),
    "Gen Z": ("2015-01-01", "2030-12-31")
}

# Segment data by generation for age 18-25
def get_generation(year):
    if 1946 <= year <= 1963:
        return "Silent Generation"
    elif 1964 <= year <= 1982:
        return "Baby Boomers"
    elif 1983 <= year <= 1998:
        return "Gen X"
    elif 1999 <= year <= 2014:
        return "Millennials"
    elif 2015 <= year <= 2030:
        return "Gen Z"
    else:
        return None

df['generation'] = df['observation_date'].dt.year.apply(get_generation)

# Filter out rows where generation is None
df = df[df['generation'].notna()]

# Calculate average unemployment rate per generation
generation_avg = df.groupby('generation')['unemployment_rate'].mean().reset_index()

# Plotting the data
plt.figure(figsize=(12, 6))
plt.bar(generation_avg['generation'], generation_avg['unemployment_rate'], color=['gray', 'blue', 'green', 'orange', 'red'])
plt.title('Average Youth Unemployment Rate (Ages 18-25) by Generation in Kenya')
plt.xlabel('Generation')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Display the plot
plt.show()
```

- **Millennials (1999-2014):** The average rate is slightly higher at 6.8%.
- **Gen X (1983-1998):** The average rate is the lowest among the three at 6.5%.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'SLEMPTOTLSPZSKEN- Employment to Population Ratio for Kenya, Percent, Annual, Not Seasonally Adjusted - FRED Graph.csv'
df1 = pd.read_csv(file_path, skiprows=1)

# Rename the columns to more meaningful names
df1.columns = ['Date', 'Employment_to_Population_Ratio']

# Convert the Date column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])

# Add a year column
df1['Year'] = df1['Date'].dt.year

# Define the generations
generations = {
    'Silent Generation': (1928, 1945),
    'Baby Boomers': (1946, 1964),
    'Gen X': (1965, 1980),
    'Millennials': (1981, 1996),
    'Gen Z': (1997, 2012),
    'Gen Alpha': (2013, 2025)
}

def get_working_age_range(birth_year):
    return birth_year + 18, birth_year + 25

# Filter data for each generation for the age range 18-25 years
gen_data = {}

for gen, (start_year, end_year) in generations.items():
    working_ages = [get_working_age_range(year) for year in range(start_year, end_year + 1)]
    working_ages = [age for age_range in working_ages for age in range(age_range[0], age_range[1] + 1)]
    
    gen_data[gen] = df1[df1['Year'].isin(working_ages)]

# Calculate statistics for each generation
gen_stats = {gen: gen_data[gen]['Employment_to_Population_Ratio'].describe() for gen in gen_data}

# # Calculate statistics for each generation
# gen_stats = {gen: gen_data[gen]['Employment_to_Population_Ratio'].describe() for gen in gen_data}

# # Display the statistics
# for gen in gen_stats:
#     print(f"Statistics for {gen}:")
#     print(gen_stats[gen])
#     print("\n")

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set a seaborn style
sns.set(style="whitegrid")

# Custom palette for more vibrant colors
custom_palette = ['#4c72b0', '#55a868', '#c44e52', '#8172b2']

# Define a function for plotting population distribution with creative styling
def plot_population_distribution(df, palette):
    columns = ['No_Education_-_Male', 'Primary_-_Male', 'Secondary_-_Male', 'Tertiary_-_Male',
               'No_Education_-_Female', 'Primary_-_Female', 'Secondary_-_Female', 'Tertiary_-_Female']
    
    # Aggregate total counts across both genders
    total_distribution = df[columns].sum(axis=0)
    education_levels = [col.split('_-_')[0] for col in columns]
    
    plt.figure(figsize=(12, 6))  # Adjusted figure size
    sns.barplot(x=education_levels, y=total_distribution, palette=palette, ci=None)
    plt.title('Population Distribution by Education Level', fontsize=18, fontweight='bold', color='#333333', loc='center', pad=20)  # Title with custom font and padding
    plt.xlabel('Education Level', fontsize=14, fontweight='bold', color='#555555')  # X-axis label with custom font
    plt.ylabel('Total Count', fontsize=14, fontweight='bold', color='#555555')  # Y-axis label with custom font
    plt.xticks(rotation=45, fontsize=12, fontweight='bold', color='#555555')  # X-axis tick labels with custom font and rotation
    plt.yticks(fontsize=12, fontweight='bold', color='#555555')  # Y-axis tick labels with custom font
    
    sns.despine(left=True)  # Remove left spine
    
    # Adding grid lines with custom styling
    plt.grid(axis='y', linestyle='--', linewidth=0.5, color='#aaaaaa')

    # Adding artistic elements like annotation
    for p in plt.gca().patches:
        plt.gca().annotate(f'{int(p.get_height()):,}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                           ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), 
                           textcoords='offset points')
    
    plt.tight_layout()
    plt.show()

# Plotting population distribution for Male and Female with the custom palette
plot_population_distribution(df3, custom_palette)
```
