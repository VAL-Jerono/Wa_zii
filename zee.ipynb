- To focus on the youth unemployment rate within the 18-25 age group for each generation, we'll need to adjust the generation ranges to reflect when they reached the working age of 18. Then, we'll segment the data accordingly and plot the unemployment rates for each generation during their 18-25 age period.

Steps:
- Adjust the date ranges for each generation to reflect the age range of 18-25 years.
- Segment the unemployment data according to these adjusted generational ranges.
- Plot the data to visualize the differences in unemployment rates across generations for the 18-25 age group.

```python
# Convert observation_date to datetime
df['observation_date'] = pd.to_datetime(df['observation_date'])

# Define adjusted generation ranges to reflect age 18-25 years
generations = {
    "Silent Generation": ("1946-01-01", "1963-12-31"),
    "Baby Boomers": ("1964-01-01", "1982-12-31"),
    "Gen X": ("1983-01-01", "1998-12-31"),
    "Millennials": ("1999-01-01", "2014-12-31"),
    "Gen Z": ("2015-01-01", "2030-12-31")
}

# Segment data by generation for age 18-25
def get_generation(year):
    if 1946 <= year <= 1963:
        return "Silent Generation"
    elif 1964 <= year <= 1982:
        return "Baby Boomers"
    elif 1983 <= year <= 1998:
        return "Gen X"
    elif 1999 <= year <= 2014:
        return "Millennials"
    elif 2015 <= year <= 2030:
        return "Gen Z"
    else:
        return None

df['generation'] = df['observation_date'].dt.year.apply(get_generation)

# Filter out rows where generation is None
df = df[df['generation'].notna()]

# Calculate average unemployment rate per generation
generation_avg = df.groupby('generation')['unemployment_rate'].mean().reset_index()

# Plotting the data
plt.figure(figsize=(12, 6))
plt.bar(generation_avg['generation'], generation_avg['unemployment_rate'], color=['gray', 'blue', 'green', 'orange', 'red'])
plt.title('Average Youth Unemployment Rate (Ages 18-25) by Generation in Kenya')
plt.xlabel('Generation')
plt.ylabel('Average Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Display the plot
plt.show()
```

**Generational Analysis:**

- **Gen Z (2015-2030):** The average youth unemployment rate is the highest at 10.5%.
- **Millennials (1999-2014):** The average rate is slightly higher at 6.8%.
- **Gen X (1983-1998):** The average rate is the lowest among the three at 6.5%.

#### Employment Rate

```python
import pandas as pd

# Load the CSV file
file_path = 'SLEMPTOTLSPZSKEN- Employment to Population Ratio for Kenya, Percent, Annual, Not Seasonally Adjusted - FRED Graph.csv'
df1 = pd.read_csv(file_path, skiprows=1)

# Rename the columns to more meaningful names
df1.columns = ['Date', 'Employment_to_Population_Ratio']

# Convert the Date column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])

# Add a year column
df1['Year'] = df1['Date'].dt.year

# Display the cleaned dataframe
print(df1.head())
```

```python
generations = {
    'Silent Generation': (1928, 1945),
    'Baby Boomers': (1946, 1964),
    'Gen X': (1965, 1980),
    'Millennials': (1981, 1996),
    'Gen Z': (1997, 2012),
    'Gen Alpha': (2013, 2025)
}

def get_working_age_range(birth_year):
    return birth_year + 18, birth_year + 25

# Filter data for each generation for the age range 18-25 years
gen_data = {}

for gen, (start_year, end_year) in generations.items():
    working_ages = [get_working_age_range(year) for year in range(start_year, end_year + 1)]
    working_ages = [age for age_range in working_ages for age in range(age_range[0], age_range[1] + 1)]
    
    gen_data[gen] = df1[df1['Year'].isin(working_ages)]

# Display the filtered data for Gen Z as an example
print(gen_data['Gen Z'].head())
```

```python
# Calculate statistics for each generation
gen_stats = {gen: gen_data[gen]['Employment_to_Population_Ratio'].describe() for gen in gen_data}

# Display the statistics
for gen in gen_stats:
    print(f"Statistics for {gen}:")
    print(gen_stats[gen])
    print("\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter generations from Gen X to Gen Z
filtered_gens = ['Gen X', 'Millennials', 'Gen Z']

# Set the style for the plots
sns.set(style="whitegrid")

# Plot the employment to population ratio for Gen X, Millennials, and Gen Z
plt.figure(figsize=(14, 8))

for gen in filtered_gens:
    plt.plot(gen_data[gen]['Year'], gen_data[gen]['Employment_to_Population_Ratio'], marker='o', label=gen)

plt.title('Employment to Population Ratio (18-25 age group) for Gen X, Millennials, and Gen Z')
plt.xlabel('Year')
plt.ylabel('Employment to Population Ratio (%)')
plt.legend()
plt.grid(True)
plt.show()
```
