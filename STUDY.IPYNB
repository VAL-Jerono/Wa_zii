*Steps for Analysis:*

2. Data Preparation:

```python
# Load the population by age, sex, and educational attainment data
population_age_sex_education = pd.read_csv("population_by_age_sex_and_educationa_attainment_in_2020.csv")

# Display the columns and first few rows to understand the structure
print("\nColumns and first few rows in population_by_age_sex_and_educational_attainment_in_2020.csv:")
print(population_age_sex_education.columns)
print(population_age_sex_education.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
file_path = "population_by_age_sex_and_educationa_attainment_in_2020.csv"
data = pd.read_csv(file_path)

# Correct negative values
data.iloc[:, 1:-1] = data.iloc[:, 1:-1].abs()

# Define generations
generations = {
    "Silent Generation": ["75-79", "80-84", "85-89", "90-94", "95-99", "100+"],
    "Baby Boomers": ["55-59", "60-64", "65-69", "70-74"],
    "Gen X": ["40-44", "45-49", "50-54"],
    "Millennials": ["25-29", "30-34", "35-39"],
    "Gen Z": ["5-9", "10-14", "15-19", "20-24"],
}

# Assign colours to generations
generation_colours = {
    "Silent Generation": "grey",
    "Baby Boomers": "green",
    "Gen X": "orange",
    "Millennials": "purple",
    "Gen Z": "blue",
}

# Add a generation column to the data
data["Generation"] = "Others"
for gen, brackets in generations.items():
    data.loc[data["Age_Bracket"].isin(brackets), "Generation"] = gen

# Calculate total educational attainment for each age bracket
data["Total_Attainment"] = data.iloc[:, 1:9].sum(axis=1)

# Plot total educational attainment by age bracket
plt.figure(figsize=(14, 8))
for gen, gen_data in data.groupby("Generation"):
    plt.bar(
        gen_data["Age_Bracket"],
        gen_data["Total_Attainment"],
        color=generation_colours.get(gen, "grey"),
        label=gen,
        alpha=0.8,
    )

# Add titles and labels
plt.title("Educational Attainment by Age Bracket and Generation (2020)", fontsize=16)
plt.xlabel("Age Bracket", fontsize=14)
plt.ylabel("Total Educational Attainment", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(title="Generations", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()



# Define Generation Z age brackets (1997–2012 born, 2020 ages: 8–23)
gen_z_brackets = ["5-9", "10-14", "15-19", "20-24"]
data["Is_Gen_Z"] = data["Age_Bracket"].isin(gen_z_brackets)

# Calculate total educational attainment for each age bracket
data["Total_Attainment"] = data.iloc[:, 1:9].sum(axis=1)

# Plot total educational attainment by age bracket
plt.figure(figsize=(12, 6))
for is_gen_z, group_data in data.groupby("Is_Gen_Z"):
    plt.bar(
        group_data["Age_Bracket"],
        group_data["Total_Attainment"],
        color="blue" if is_gen_z else "grey",
        label="Gen Z" if is_gen_z else "Others",
        alpha=0.7,
    )

# Add titles and labels
plt.title("Educational Attainment by Age Bracket (Gen Z Highlighted)", fontsize=16)
plt.xlabel("Age Bracket", fontsize=14)
plt.ylabel("Total Educational Attainment", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend(fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()



# Calculate total attainment for males and females
data["Male_Attainment"] = data.iloc[:, 1:5].sum(axis=1)
data["Female_Attainment"] = data.iloc[:, 5:9].sum(axis=1)

# Plot total educational attainment by age bracket for males and females
plt.figure(figsize=(16, 8))
x = np.arange(len(data))  # X positions for the bars
bar_width = 0.35  # Reduced bar width for clarity

# Create empty lists for legend handles and labels
legend_handles = []
legend_labels = []

# Plot bars for each gender and generation
for gen, gen_data in data.groupby("Generation"):
    indices = gen_data.index
    male_bar = plt.bar(
        x[indices] - bar_width / 2,
        gen_data.loc[indices, "Male_Attainment"],
        width=bar_width,
        color=generation_colours.get(gen, "grey"),
        alpha=0.7,
    )
    female_bar = plt.bar(
        x[indices] + bar_width / 2,
        gen_data.loc[indices, "Female_Attainment"],
        width=bar_width,
        color=generation_colours.get(gen, "grey"),
        alpha=0.5,
    )
    
    # Only add the legend once for each generation and gender
    if f"{gen} - Male" not in legend_labels:
        legend_handles.append(male_bar)
        legend_labels.append(f"{gen} - Male")
    if f"{gen} - Female" not in legend_labels:
        legend_handles.append(female_bar)
        legend_labels.append(f"{gen} - Female")

# Add titles, labels, and legend
plt.title("Educational Attainment by Age Bracket, Gender, and Generation (2020)", fontsize=16)
plt.xlabel("Age Bracket", fontsize=14)
plt.ylabel("Total Educational Attainment", fontsize=14)
plt.xticks(x, data["Age_Bracket"], rotation=45, fontsize=12)

# Move the legend outside of the plot to avoid overlap with bars
plt.legend(handles=legend_handles, labels=legend_labels, title="Generations & Gender", fontsize=12, loc="upper left", bbox_to_anchor=(1, 1))

plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()  # Ensure everything fits within the figure
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
population_census = pd.read_excel("population census 2019.xlsx")

# Define generations and their respective age groups
generations = {
    "Silent Generation": ['T_75_79', 'T_80_84', 'T_85_89', 'T_90_94', 'T_95_99', 'T_100Plus'],
    "Baby Boomers": ['T_55_59', 'T_60_64', 'T_65_69', 'T_70_74'],
    "Gen X": ['T_40_44', 'T_45_49', 'T_50_54'],
    "Millennials": ['T_25_29', 'T_30_34', 'T_35_39'],
    "Gen Z": ['T_00_04', 'T_05_09', 'T_10_14', 'T_15_19'],
    "Gen Alpha": ['T_20_24']
}

# Calculate the total population per generation
generation_totals = {}
for gen, age_groups in generations.items():
    generation_totals[gen] = population_census[age_groups].sum(axis=1).sum()

# Total Population
total_population = population_census['T_TL'].sum()

# Print total population and Generation proportions
print(f"Total Population: {total_population}")
for gen, total in generation_totals.items():
    percentage = (total / total_population) * 100
    print(f"{gen} Population: {total} ({percentage:.2f}%)")

# Visualization 1: Total Population by Generation (Bar Chart)
plt.figure(figsize=(10, 6))
plt.bar(generation_totals.keys(), generation_totals.values(), color=['grey', 'green', 'orange', 'purple', 'blue', 'cyan'], edgecolor='black')
plt.title('Total Population by Generation (2019)', fontsize=16)
plt.xlabel('Generation', fontsize=14)
plt.ylabel('Population', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Visualization 2: Population by Generation (Pie Chart)
labels = generation_totals.keys()
sizes = generation_totals.values()
colors = ['grey', 'green', 'orange', 'purple', 'blue', 'cyan']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
plt.title('Total Population Distribution by Generation (2019)', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()


```

```python
# Ensure the 'Value' column is numeric, setting errors='coerce' to convert invalid values to NaN
qeducation_data['Value'] = pd.to_numeric(qeducation_data['Value'], errors='coerce')

# Check for missing data
missing_data = qeducation_data.isnull().sum()

# Handle missing values (e.g., by filling with the forward fill method)
qeducation_data['Value'].fillna(method='ffill', inplace=True)

# Summary statistics of the 'Value' column after cleaning
value_summary = qeducation_data['Value'].describe()

# Group the data by 'Year' to understand trends over time
yearly_data = qeducation_data.groupby('Year')['Value'].sum()

# Check for the overall trend (growth or stagnation)
growth_rate = yearly_data.pct_change().mean()

# Output the results
missing_data, value_summary, growth_rate
```

```python
import matplotlib.pyplot as plt

# Plotting the trends over time
plt.figure(figsize=(10, 6))
plt.plot(yearly_data.index, yearly_data.values, marker='o', linestyle='-', color='b', label='Enrollment/Completion')

# Adding labels and title
plt.title("Trends in Education Outcomes (Kenya) Over Time", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Enrollment/Completion Count", fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.legend()
plt.show()
```
